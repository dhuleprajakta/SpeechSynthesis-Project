```python
from transformers import pipeline
import numpy as np
import scipy.io.wavfile as wav
import os

# Initialize Text-to-Speech pipeline
synthesizer = pipeline("text-to-speech", model="suno/bark-small")

def generate_speech(user_input):
    # Generate speech using the Bark model
    speech = synthesizer(user_input, forward_params={"do_sample": True})

    # Get sampling rate and audio
    sampling_rate = int(speech["sampling_rate"])
    audio = speech["audio"]

    # Reshape and convert to int16
    audio = np.squeeze(audio)
    audio_int16 = (audio * 32767).astype(np.int16)

    # Save as wav file
    output_filename = "bark_out.wav"
    wav.write(output_filename, rate=sampling_rate, data=audio_int16)

    return output_filename

# Example input
user_input = input("Enter the text you want to convert to speech: ")

if user_input.strip() != "":
    print("Generating speech... üé∂")
    output_filename = generate_speech(user_input)
    print(f"Speech generated successfully! üéâ Saved as {output_filename}")
else:
    print("‚ö†Ô∏è Please enter some text.")
```
